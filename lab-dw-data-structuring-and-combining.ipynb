## Challenge 1: Combining Data (Option 1)

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the following links:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2_clean.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3_clean.csv

Combine the data from the three dataframes into a single dataframe, named "customer_data", using appropriate merging, concatenating, and joining techniques.

Verify that the customer_data dataframe contains all the rows and columns from the three original dataframes.

```python
import pandas as pd

data_1 = pd.read_csv('file1.csv')
data_2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2_clean.csv')
data_3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3_clean.csv')

#file2 and file3 were not cleaned, therefore following cleaning measures are nessesary

def lowercase_column_names(df):
    # Convert column names to lowercase
    df.columns = [col.lower() for col in df.columns]
    
lowercase_column_names(data_2)
lowercase_column_names(data_3)

data_2 = data_2.rename(columns=lambda x: x.replace(' ', '_'))
data_3 = data_3.rename(columns=lambda x: x.replace(' ', '_'))

data_2 = data_2.drop(0, axis=0)
data_3 = data_3.drop(0, axis=0)

data_2 = data_2.rename(columns = {'st':'state'})
data_3 = data_3.rename(columns = {'st':'state'})

display(data_1.head())
display(data_2.head())
display(data_3.head())

```

```python
def compare_columns(df1, df2, df3):
    if df1.columns.equals(df2.columns) and df2.columns.equals(df3.columns):
        if df1.columns.to_list() == df2.columns.to_list() == df3.columns.to_list():
            print("All three DataFrames have the same column names and order.")
        else:
            print("The column names are the same, but the order is different.")
    else:
        print("The column names are not the same in all three DataFrames.")
        
        # Print or write unmatched columns
        unmatched_df1 = df1.columns.difference(df2.columns.union(df3.columns))
        unmatched_df2 = df2.columns.difference(df1.columns.union(df3.columns))
        unmatched_df3 = df3.columns.difference(df1.columns.union(df2.columns))

        print("Unmatched columns in DataFrame 1:", unmatched_df1.to_list())
        print("Unmatched columns in DataFrame 2:", unmatched_df2.to_list())
        print("Unmatched columns in DataFrame 3:", unmatched_df3.to_list())

compare_columns(data_1, data_2, data_3)
```

```python
def standardize_columns(df):
    # Replace values in 'gender' column
    df['gender'] = df['gender'].replace({'Femal': 'F', 'female': 'F', 'Male': 'M'})
    
    # Replace values in 'state' column
    df['state'] = df['state'].replace({'AZ': 'Arizona', 'Cali': 'California', 'WA': 'Washington'})
    
    # Replace values in 'education' column
    df['education'] = df['education'].replace({'Bachelors': 'Bachelor'})
    
    # Remove '%' from 'customer_lifetime_value' column
    df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '')
    
    # Replace values in 'vehicle_class' column
    df['vehicle_class'] = df['vehicle_class'].replace({'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury'})
    
    return df

data_2 = standardize_columns(data_2)
data_3 = standardize_columns(data_3)
```

```python
#transform to numeric
data_2['customer_lifetime_value'] = pd.to_numeric(data_2['customer_lifetime_value'], errors = 'coerce')
data_3['customer_lifetime_value'] = pd.to_numeric(data_3['customer_lifetime_value'], errors = 'coerce')

#fix the format
data_2['number_of_open_complaints'] = data_2['number_of_open_complaints'].apply(lambda x: int(str(x).split('/')[1]) if '/' in str(x) else x)
data_3['number_of_open_complaints'] = data_3['number_of_open_complaints'].apply(lambda x: int(str(x).split('/')[1]) if '/' in str(x) else x)
```

```python
data_2 = data_2.dropna(how = 'all')
nulls_2 = data_2.isnull().sum()

data_3 = data_3.dropna(how = 'all')
nulls_3 = data_3.isnull().sum()

print(nulls_2)
print(nulls_3)
```

```python
# Fill null values in customer_lifetime_value with the mean
mean_value_2 = data_2['customer_lifetime_value'].mean()
data_2['customer_lifetime_value'].fillna(mean_value_2, inplace=True)

mean_value_3 = data_3['customer_lifetime_value'].mean()
data_3['customer_lifetime_value'].fillna(mean_value_3, inplace=True)

nulls_2 = data_2.isnull().sum()
nulls_3 = data_3.isnull().sum()

print(nulls_2)
print(nulls_3)
```

```python
# Fill null values in 'gender' with 'U'
data_2['gender'].fillna('U', inplace=True)
data_3['gender'].fillna('U', inplace=True)

nulls_2 = data_2.isnull().sum()
nulls_3 = data_3.isnull().sum()

print(nulls_2)
print(nulls_3)
```

```python
# Check for duplicate rows
duplicate_rows_2 = data_2[data_2.duplicated()]
duplicate_rows_3 = data_3[data_3.duplicated()]

# Display the duplicate rows, if any
print("Duplicate Rows in data_2:")
display(duplicate_rows_2)

print("Duplicate Rows in data_3:")
display(duplicate_rows_3)
```

```python
data_2 = data_2.drop_duplicates()
data_3 = data_3.drop_duplicates()
```

```python
print(data_1.shape, data_2.shape, data_3.shape)
```

```python
data_2.to_csv('file2.csv', index=False)
data_3.to_csv('file3.csv', index=False)
```

```python
# Combine rows from all three DataFrames
customer_data = pd.concat([data_1, data_2, data_3], ignore_index=True)

# Print or display the combined DataFrame
print("Combined DataFrame:")
display(customer_data)
```

```python
# Check Rows
rows_check = len(data_1) + len(data_2) + len(data_3) == len(customer_data)

print("Rows Check:", rows_check)
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')
print(df.shape)

#delete unnamed column
df = df.drop(df.columns[0], axis=1)

cols = df.columns
column_names = pd.DataFrame(columns=cols)
print('Current column names:')
display(column_names.T)

print('Data types:')
display(df.dtypes)

non_numerical_columns = df.select_dtypes(include='object')
columns_to_exclude = ['effective_to_date', 'customer']
non_numerical_columns = non_numerical_columns.drop(columns=columns_to_exclude, errors='ignore')

df = df.drop(df.columns[0])
unique_values = {}
for column in non_numerical_columns.columns:
    unique_values[column] = non_numerical_columns[column].unique()

# Display unique values
for column, values in unique_values.items():
    print(f"\nUnique values for {column}:\n{values}")
```

```python
#check NaNsb
nulls = df.isnull().sum()
print(nulls)
```

```python
#check duplicates
duplicate_rows = df[df.duplicated()]
display(duplicate_rows)
```

```python
# Round up to the whole number
df['months_since_last_claim'] = df['months_since_last_claim'].apply(np.ceil)
df['number_of_open_complaints'] = df['number_of_open_complaints'].apply(np.ceil)
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
summary_table = df.pivot_table(values='monthly_premium_auto', index='sales_channel', aggfunc='sum').round(2)
summary_table['percentage'] = (summary_table['monthly_premium_auto'] / summary_table['monthly_premium_auto'].sum() * 100).round(2)
# Display the summary table
print("Summary Table:")
display(summary_table.reset_index())
```

As we can see from the summary table, the most successful sales channel is 'Agent' with 38% of the total revenue. The least total revenue 151511 was calected through the 'Web'. The channels 'Branch' and 'Call Center' have 27.63% and 19.47% correspondingly.

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
pivot_result = df.pivot_table(
    values='customer_lifetime_value',
    index=['education','gender'],
    aggfunc='mean'
).round(2)

pivot_result.columns = ['average_customer_lifetime_value']

display(pivot_result)

pivot_result = df.pivot_table(
    values='customer_lifetime_value',
    index=['gender','education'],
    aggfunc='mean'
).round(2)

pivot_result.columns = ['average_customer_lifetime_value']

display(pivot_result)
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
pivot_result = df.pivot_table(
    values='number_of_open_complaints',
    index='policy_type',
    columns='month',
    aggfunc='sum',
    fill_value=0
)

print("Pivot Result:")
display(pivot_result)
```
